```python
labels='India','US','China','South Africa'
sizes=(41,39,27,13)
explode=(0.1,0,0,0)
colors=('orange','blue','red','green')
plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%',startangle=90, shadow=True, colors=colors)
plt.axis('equal')
plt.show()
```
